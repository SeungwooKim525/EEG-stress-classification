```python
#데이터 분석
import scipy.io
from scipy.io import loadmat

data= loadmat('DREAMER.mat')

print(data.keys()) #key check
```

```python
#main dataset:DREAMER
#구조분석
dreamer=data['DREAMER']

print(type(dreamer))

print(dreamer.shape)

if hasattr(dreamer.dtype,'names'): #dreamer필드 안의 세부필드확인
    print(dreamer.dtype.names)
```

```python
dreamer_struct=dreamer[0,0]  #배열 해제

main_data=dreamer_struct['Data'] #main data 
print(type(main_data))
print(main_data.shape)

no_subjects = dreamer_struct['noOfSubjects'] #메타데이터
print(f"\n참가자 수 ('noOfSubjects'): {no_subjects}")

eeg_rate = dreamer_struct['EEG_SamplingRate'] #샘플링 속도
print(f"\nEEG 샘플링 속도: {eeg_rate}")
```

```python
subject = main_data[0, 0] #피험자

subject1= subject[0, 0]
#print(subject1)
if hasattr(subject1.dtype, 'names'):
    print(f"\n참가자1 데이터의 내부 필드 이름: {subject1.dtype.names}")
```

```python
#eeg
eeg_container= subject1['EEG'][0,0]
print(f"EEG 신호 데이터 모양: {eeg_container.shape}") 
```

```python
#eeg 세부필드 확인
if hasattr(eeg_container.dtype, 'names'):
    print(f"\n숨겨진 EEG 컨테이너 내부 필드 이름: {eeg_container.dtype.names}")
```

```python
baseline=eeg_container['baseline']
print(f"baseline1: {baseline[0]}")

stimuli=eeg_container['stimuli']
print(f"stimuli1: {stimuli[0]}")
```
